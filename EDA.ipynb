## 5. Exploratory Data Analysis (EDA)

### 5.1 Sales Overview

```python
df["Sales"].sum()
```

### 📌 Business Insight

- The supermarket generated significant overall sales during the recorded period.
- The average sales value represents the typical amount spent by customers.
- The highest and lowest sales indicate the range of customer spending behavior.

```python
df.groupby("City")["Sales"].sum()
```

```python
df["Payment"].value_counts()
```

```python
df["Payment"].value_counts().plot(kind="bar")
plt.title("Payment Methods Used")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.show()
```

### 5.2 Product Line Analysis

```python
product_sales = df.groupby("Product line")["Sales"].sum()
product_sales
```

```python
product_sales.plot(kind="bar", figsize=(10,5))

plt.title("Total Sales by Product Line")

plt.xlabel("Product Line")

plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

### 5.3 Branch-wise Sales Analysis

```python
branch_sales = df.groupby("Branch")["Sales"].sum()

branch_sales
```

```python
branch_sales.plot(kind="bar", figsize=(8,5))

plt.title("Total Sales by Branch")

plt.xlabel("Branch")

plt.ylabel("Sales")

plt.show()
```

### 5.4 City-wise Sales Analysis

```python
city_sales = df.groupby("City")["Sales"].sum()

city_sales
```

```python
city_sales.plot(kind="bar", figsize=(8,5))

plt.title("Total Sales by City")

plt.xlabel("City")

plt.ylabel("Sales")

plt.show()
```

### 5.5 Payment Method Analysis

```python
payment_count = df["Payment"].value_counts()

payment_count
```

```python
payment_count.plot(kind="bar", figsize=(8,5))

plt.title("Preferred Payment Methods")

plt.xlabel("Payment Method")

plt.ylabel("Number of Customers")

plt.show()
```

### 5.6 Customer Rating Analysis

```python
df["Rating"].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Rating"], bins=10)

plt.title("Distribution of Customer Ratings")

plt.xlabel("Rating")

plt.ylabel("Number of Customers")

plt.show()
```

### 5.7 Sales Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=20)

plt.title("Distribution of Sales")

plt.xlabel("Sales Amount")

plt.ylabel("Number of Customers")

plt.show()
```

### 5.8 Quantity Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["Quantity"], bins=10)

plt.title("Quantity Purchased Distribution")

plt.xlabel("Quantity")

plt.ylabel("Number of Customers")

plt.show()
```

### 5.9 Corelation Analysis

```python
plt.figure(figsize=(10,6))

correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df, x="Product line")

plt.xticks(rotation=45)

plt.title("Number of Purchases by Product Line")

plt.show()
```

### 5.10 Gender Distribution 

```python
plt.figure(figsize=(6,5))

sns.countplot(data=df, x="Gender")

plt.title("Gender Distribution")

plt.show()
```

### 5.11 Product Line Rating Analysis

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=df, x="Product line", y="Rating")

plt.xticks(rotation=45)

plt.title("Customer Ratings by Product Line")

plt.show()
```
